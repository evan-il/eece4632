```python
import cv2
import os
import numpy as np
from drawcube3 import project_cube
```

```python
FRAMES_DIR    = "output_frames"   # folder of input frames
OUTPUT_VIDEO  = "cube_output.mp4"
FPS           = 30.0
REAL_RADIUS   = 50.0              # mm
Z_OFFSET      = 0.0
```

```python
frame_files = sorted(
    f for f in os.listdir(FRAMES_DIR)
    if f.lower().endswith((".png", ".jpg", ".jpeg"))
)
print(f"Found {len(frame_files)} frames")
```

```python
sample = cv2.imread(os.path.join(FRAMES_DIR, frame_files[0]))
h, w   = sample.shape[:2]
writer = cv2.VideoWriter(OUTPUT_VIDEO, cv2.VideoWriter_fourcc(*'mp4v'), FPS, (w, h))

for i, fname in enumerate(frame_files):
    path = os.path.join(FRAMES_DIR, fname)
    try:
        frame = project_cube(path, REAL_RADIUS, Z_OFFSET)
    except (ValueError, RuntimeError):
        frame = cv2.imread(path)
    writer.write(frame)
    if (i + 1) % 50 == 0:
        print(f"  {i+1}/{len(frame_files)}")

writer.release()
print(f"Saved: {OUTPUT_VIDEO}")
```
